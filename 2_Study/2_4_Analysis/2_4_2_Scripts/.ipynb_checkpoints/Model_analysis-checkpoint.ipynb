```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('m2_trace.csv')
fig, ax = plt.subplots(figsize = (10, 6))

ax1 = sns.distplot(df['v_C(coherency, Treatment("low"))[T.high]'],rug=False,hist=False)
ax2 = sns.distplot(df['v_cpp_slope'],rug=False,hist=False)
ax3 = sns.distplot(df['v_cpp_slope:C(coherency, Treatment("low"))[T.high]'],rug=False,hist=False)
ax4 = sns.distplot(df['t_C(prioritization, Treatment("no"))[T.yes]'],rug=False,hist=False)
plt.legend(["C","CPP","C:CPP","A"])
plt.savefig("m2_A.png")
```
